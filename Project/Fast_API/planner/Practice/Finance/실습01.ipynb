# 시총 TOP10 업종명, 업종코드, 종가 가져오기
* 기준일자 : 20230403
* krx API 이용

* df = stock.get_market_ohlcv("20220720", "20220810", "005930")

## 방법1) 기준일자 : 20230403

```python
df = stock.get_market_cap("20230403")[:10]
df = df[['종가']]
df = df.reset_index()

temp_list = []
for ticker in df['티커']:
    티커명 = stock.get_market_ticker_name(ticker)
    temp_list.append(티커명)
df['티커명'] = temp_list
df
```

## 방법2) 기준일자 : 20230403 ~ 20230425

```python
# ohlcv_df = stock.get_market_ohlcv("20230403", "20230425", "005490")
# ohlcv_df = ohlcv_df.iloc[:, 3:4]
# ohlcv_df.columns = ['삼성']
# ohlcv_df
```

```python
idx_list = stock.get_market_cap("20230403")[:10].index
ticker_name_list = []
for ticker in df['티커']:
    티커명 = stock.get_market_ticker_name(ticker)
    print(ticker, 티커명)
    ticker_name_list.append(티커명)
ticker_name_list    
```

```python
df_list = []
for ticker in idx_list:
    ohlcv_df = stock.get_market_ohlcv("20230403", "20230425", ticker)
    ohlcv_df = ohlcv_df.iloc[:, 3:4]
    df_list.append(ohlcv_df)
    
tot_df = pd.concat(df_list, axis=1)
tot_df.columns = ticker_name_list  
tot_df
```

# 상위 종목 지수화
* 수익률
* ret_df

```python
ret_df = tot_df.pct_change(1)
ret_df
```

# 누적 수익률 차트

```python
sns.set()

#-------------------- 차트 관련 속성 (한글처리, 그리드) -----------
#plt.rc('font', family='NanumGothicOTF') # For MacOS
plt.rcParams['font.family']= 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(20,6))
plt.title('누적 수익률 비교')
plt.plot(ret_df.cumsum() , label=ret_df.columns.values)
plt.legend(loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.title('누적 수익률 비교')
for col in ret_df.columns:
    plt.plot(ret_df.index  , ret_df[col].cumsum(), label=col)
plt.legend(loc='upper left')
plt.show()
```
