# 6차 학습
* 피쳐 선택, 로그 스케일, 아웃라이어 처리, 피쳐 임포턴스 후에 피쳐선택,  파라미터 튜닝

```python
y = np.log1p(train["SalePrice"])
X = np.log1p(train.drop("SalePrice", axis=1))
```

```python
def my_objective_GBM(trial, data=X, target=y):
    
    X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2,random_state=42)
    param = {
        
        'random_state': 1,
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.01,0.1),
        'n_estimators': trial.suggest_int('n_estimators', 50,500),  #100
        
        'min_impurity_decrease' : trial.suggest_int('min_samples_split', 1.0, 5.0),
        # 'min_samples_split' : trial.suggest_float('min_samples_split', 0.0, 2.0),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1.0, 5.0),  #reg_alpha=0.0
        
        # 'max_depth': trial.suggest_categorical('max_depth', [1,30]),
        #'colsample_bytree': trial.suggest_categorical('colsample_bytree', [0.3,0.4,0.5,0.6,0.7,0.8,0.9, 1.0]),
        #'subsample': trial.suggest_categorical('subsample', [0.4,0.5,0.6,0.7,0.8,1.0]),
        #'num_leaves' : trial.suggest_int('num_leaves', 1, 1000),
        
        #'cat_smooth' : trial.suggest_int('min_data_per_groups', 1, 100)
    }
    gbr_modified = GradientBoostingRegressor(**param)  
    
    gbr_modified.fit(X_train,y_train)
              # ,eval_set=[(X_test,y_test)]
              # ,early_stopping_rounds=10,verbose=False)
    pred = gbr_modified.predict(X_test)
    
    rmse_modified = mean_squared_error(y_test, pred, squared=False)
    return rmse_modified
```

```python

study = optuna.create_study(direction='minimize')


# y_train = np.expm1(y_train_log)
# y_test  = np.expm1(y_test_log)


# my_objective_LGBM(trial):
study.optimize(my_objective_GBM, n_trials=50)

# def my_objective_LGBM(trial, data=X_train, target=y_train):
# study.optimize(lambda trial : my_objective_LGBM(trial, y_train,y_test), n_trials=50)
```
