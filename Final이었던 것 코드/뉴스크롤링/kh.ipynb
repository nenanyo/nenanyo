# 검색 본문 크롤링

```python
def craw_content(search):
    news_list = []
    for pageno in range(1, 500, 1):
        interval = round(random.uniform(0.2, 1.2), 2)
        time.sleep(interval)
        #------------------------------------------------
        url = "https://www.theinvestor.co.kr/search_news.php?q=" + search + "&np=" + str(pageno)
        # print(url)

        response = requests.get(url)
        # print(response)

        if response.status_code == 200:
            html = response.text
            soup = BeautifulSoup(html, 'html.parser')
            div_list = soup.select("#container > div.article_list_left > div > ul > li")
            # print(div_list)
            
            for i, li_tag in enumerate(div_list):

                dict = {}
                title = li_tag.select_one('#container > div.article_list_left > div > ul > li > p.title > a').text
                rdate = li_tag.select_one('#container > div.article_list_left > div > ul > li > span').text
                href = li_tag.select_one('#container > div.article_list_left > div > ul > li > p.title > a').get("href")
                # cont = li_tag.select_one('#container > div.article_list_left > div > ul > li > p.txt > a').text
               

                dict['key_title'] = title
                dict['key_rdate'] = rdate
                dict['key_href'] = href
                # dict['key_cont'] = cont

                # news_list.append(dict)
                # print(href)
                response_sub = requests.get(href)
                # print(response_sub)
                if response_sub.status_code == 200:
                    interval = round(random.uniform(0.2, 1.2), 2)
                    time.sleep(interval)

                    html_sub = response_sub.text
                    html_soup = BeautifulSoup(html_sub, 'html.parser')
                    content = html_soup.select_one("#articleText").text
                    # articleText > p:nth-child(3)
                    temp = ""
                    for cc in content.rsplit("\n"):
                        if len(cc) >= 1:
                            temp += cc
                    # print(temp)
                    try:
                        dict["content"] = temp
                        news_list.append(dict)

                        # sql = "insert into craw_ytn_news(seq, title, content, cate, rdate) values (craw_ytn_news_seq.nextval,  :2, :3, :4, :5)"
                        # conn.execute(sql, (title, temp, cate, rdate))
                    except Exception as e:
                        continue
                        # trans.rollback()
                        # print(e)
                        print("에러발생")
    
    
    
        else:
            print("에러발생" + response.status_code)
            
    df = pd.DataFrame(news_list)#, columns = ["title","rdate","href","cont"])
    df.to_csv("./datasets/kh_list_profit500.csv", index=False)
    return df
```
