```python
import os
file_list_재무 = os.listdir(path_재무)
len(file_list_재무)
```

```python
file_name_재무 = []
for file in file_list_재무:
    if file.count(".") == 1: 
        name = file.split('.')[0]
        file_name_재무.append(name)
    # else:
    #     for k in range(len(file)-1,0,-1):
    #         if file[k]=='.':
    #             file_name.append(file[:k])
    #             break
```

```python
df1 = []
for 기업 in file_name_재무:
    df =pd.read_excel(f"./데이터/기업별 재무제표/재무제표 원본/재무제표/{기업}.xlsx",sheet_name='Data_bs',header=0,skiprows=[1,2])
    # df = df[['Unnamed: 2','20221231','20211231','20201231','20191231','20181231']]
    df = pd.DataFrame(df)
    df1.append(df)
len(df1)
```

```python
df2 =[]
for df in df1:
    df.columns = df.columns.astype('str')
    df = df[['Unnamed: 2','20221231','20211231','20201231','20191231','20181231']]
    df2.append(df)
```

```python
for i, df in enumerate(df2):
    df.to_csv(f"./데이터/기업별 재무제표/재무제표 원본/재무제표_csv/{file_name_재무[i]}.csv")
    
```
