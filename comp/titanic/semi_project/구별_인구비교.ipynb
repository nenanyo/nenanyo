# 지역구별 카드소비데이터 로드

```python
cddf = pd.read_csv('./집계구별 일별소비지역별 카드소비패턴.csv',encoding='cp949')
cddf.head()
```

```python
cddf = pd.DataFrame(cddf)
cddf.head()
```

```python
cddf = cddf.dropna()
```

```python
drop_idx = cddf[cddf['가맹점주소광역시도(SIDO)']!='서울'].index.tolist()
```

```python
cddf = cddf.drop(drop_idx, axis = 0)
```

```python
cddf = cddf.drop('고객주소집계구별(TOT_REG_CD)', axis = 1)
cddf.info()
```

```python
cddf['업종대분류(UPJONG_CLASS1)'] = cddf['업종대분류(UPJONG_CLASS1)'].replace(['가전/가구','가정생활/서비스'],'가정생활')
cddf['업종대분류(UPJONG_CLASS1)'] = cddf['업종대분류(UPJONG_CLASS1)'].replace('교육/학원','교육')
cddf['업종대분류(UPJONG_CLASS1)'] = cddf['업종대분류(UPJONG_CLASS1)'].replace(['스포츠/문화/레저','여행/교통'],'문화')
cddf['업종대분류(UPJONG_CLASS1)'] = cddf['업종대분류(UPJONG_CLASS1)'].replace(['주유','자동차'],'교통')
cddf['업종대분류(UPJONG_CLASS1)'] = cddf['업종대분류(UPJONG_CLASS1)'].replace(['요식/유흥','음/식료품'],'식품')
cddf['업종대분류(UPJONG_CLASS1)'] = cddf['업종대분류(UPJONG_CLASS1)'].replace(['의류/잡화','미용'],'패션')
cddf['업종대분류(UPJONG_CLASS1)'].value_counts()
```

```python
cddf['기준일자(YMD)'] = cddf['기준일자(YMD)'].astype(str)
cddf['y'] = cddf['기준일자(YMD)'].str[0:4]
cddf['m'] =cddf['기준일자(YMD)'].str[4:6]
cddf['d'] =cddf['기준일자(YMD)'].str[6:8]

cddf.head()
```

```python
cddf['분기'] = cddf['m']
cddf['분기'] = cddf['분기'].replace(['01','02','03'],'1/4')
cddf['분기'] = cddf['분기'].replace(['04','05','06'],'2/4')
cddf['분기'] = cddf['분기'].replace(['07','08','09'],'3/4')
cddf['분기'] = cddf['분기'].replace(['10','11','12'],'4/4')
cddf.head()

```

```python
cddf['분기'] = cddf['y'] + ' ' + cddf['분기']
cddf.head()
```

```python
cddf = cddf.sort_values('분기')
cddf.head()
```

```python
cddf = cddf.drop(['y','m','d','분기'], axis = 
                1)
```

```python
cddf = cddf.drop('기준일자(YMD)', axis = 1)
cddf.head()
```
